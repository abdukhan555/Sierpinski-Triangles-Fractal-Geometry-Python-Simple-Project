```python
import matplotlib.pyplot as plt
import random as rd
import math
```

```python
x = [0, 2, 1]
y = [0, 0, math.sqrt(3)]

rx = 1
ry = 1.5

plt.scatter(x, y, s=20, c="red")
plt.scatter(rx, ry, s=20, c="green")

N = 1000
for i in range(1, N + 1):
    selection = rd.randint(0, len(x) - 1)
    rx = (rx + x[selection]) / 2
    ry = (ry + y[selection]) / 2
    plt.scatter(rx, ry, s = 1, c="black")

plt.show()
```
